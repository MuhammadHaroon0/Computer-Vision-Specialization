```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image=cv.imread("/content/drive/MyDrive/Computer Vision Masterclass/Images/people3.jpg",0)
```

```python
import dlib
```

```python
HOG_Face_Detector=dlib.get_frontal_face_detector()
```

```python
detections=HOG_Face_Detector(image,4)
```

```python
for face in detections:
  l,t,r,b=(face.left(),face.top(),face.right(),face.bottom())
  cv.rectangle(image,(l,t),(r,b),(0,255,255),2)
image #4 face detected
```

```python
CNN_Detector=dlib.cnn_face_detection_model_v1('/content/drive/MyDrive/Computer Vision Masterclass/Weights/mmod_human_face_detector.dat')
```

```python
image=cv.imread("/content/drive/MyDrive/Computer Vision Masterclass/Images/people3.jpg",0)
```

```python
cnn_detections=CNN_Detector(image,3)
```

```python
for face in cnn_detections:
  l,t,r,b,c=(face.rect.left(),face.rect.top(),face.rect.right(),face.rect.bottom(),face.confidence)
  print(c)
  cv.rectangle(image,(l,t),(r,b),(0,255,255),2)
image #4 face detected
```
