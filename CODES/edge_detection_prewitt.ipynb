```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the input image
image = cv2.imread('Mountain2.jpg')
plt.imshow(image)
plt.show()
```

```python
# Convert the input image into gray image
image = cv2.imread('Mountain2.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
plt.show()
```

```python
# Prewitt filter kernels (3x3)
prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_y = np.array([[1, -1, -1], [0, 0, 0], [-1, 1, 1]])
```

```python
# Apply Prewitt filters for horizontal and vertical gradients
prewitt_x_output = cv2.filter2D(image, -1, prewitt_x)
prewitt_y_output = cv2.filter2D(image, -1, prewitt_y)
```

```python
# Compute gradient magnitude
gradient_magnitude = np.sqrt(np.square(prewitt_x_output) + np.square(prewitt_y_output))
```

```python
# Apply thresholding for edge detection (adjust threshold as needed)
threshold = 15
edges = np.where(gradient_magnitude > threshold, 255, 0).astype(np.uint8)
```

```python
# Display original image and detected edges
cv2.imshow("Original Image", cv2.imread("Mountain2.jpg"))
cv2.imshow("Edges (Prewitt)", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
