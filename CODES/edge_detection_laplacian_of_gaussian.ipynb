```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the input image
image = cv2.imread('Mountain4.jpg')
plt.imshow(image)
plt.show()
```

```python
# Convert the input image into gray image
image = cv2.imread('Mountain4.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
plt.show()
```

```python
sigma = 1  # Adjust sigma as needed (controls smoothing and edge sharpness)
```

```python
# Apply Gaussian filter for smoothing
smoothed_image = cv2.GaussianBlur(image, (5, 5), sigma)  # Adjust kernel size as needed
```

```python
# Apply Laplacian filter
laplacian = cv2.Laplacian(smoothed_image, cv2.CV_64F)
```

```python
# Zero-crossing thresholding for edge detection
zero_crossing = np.where((laplacian[:-1, :-1] * laplacian[1:, 1:]) < 0, 255, 0).astype(np.uint8)
# Explanation: This line checks for sign change between adjacent pixels
```

```python
# Display original image and detected edges
cv2.imshow("Original Image", image)
cv2.imshow("Edges (Marr-Hildreth)", zero_crossing)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
